We find that the required dates have length 8.

```python
# We use the length to get the 
# index of the rows we need to remove

for date in data['Date']:
    if(len(date)!=8):
        index = data.index[data['Date'] == date]
print(index.values)
```

```python
# Drop the indexes which do not satisfy the condition
# Then, reset the index values

data_copy = data.drop(index.values)
#print(data_copy)
data_copy.reset_index(inplace=True)
data_copy
```

```python
# Find the number of days elapsed
elapsed = []

def find_elapsed_time(data_copy,elapsed):
    """
    Arguments: data_copy: A dataFrame variable
                elapsed: A list variable
    """
    today = datetime.today()

    dates = [datetime.strptime(x, '%d/%m/%y') for x in data_copy['Date']]
    for each_date in dates:
        elapsed.append(today - each_date)

find_elapsed_time(data_copy,elapsed)
    
data_copy['Time Elapsed'] = elapsed
data_copy['Time Elapsed']
```

```python
words = ['OpenCV', 'Python', 'Java', 'Ruby', 'Tensorflow']
buzzword_list = []
```

```python
for string in data_copy['raw']:
    if any(word in string for word in words):
        buzzword_list.append(word)
    else:
        buzzword_list.append("No")
if(len(buzzword_list )!=0):
    data_copy['Buzzwods'] = buzzword_list 
            
print(buzzword_list)
```
